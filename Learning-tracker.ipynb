## PROJECT NAME : Learning Tracker with Games

### BUSINESS PROBLEM :
programmers often struggle with consistency and motivation in their learning journey. While they practice coding concepts like loops, sets, OOP, and algorithms, they lack a structured way to track progress, measure streaks, or identify weak areas.

### PROPOSED SOLUTION:
To solve the challenge of tracking coding practice and keeping learners motivated, the project provides a Python‑based Learning Tracker with Games.

```python
from datetime import date, timedelta

concepts = []
practice_logs = {}  
streaks = {}        

def save_progress():
    with open("tracker.txt", "w") as f:
        for concept in concepts:
            logs = practice_logs.get(concept, [])
            streak = streaks.get(concept, 0)
            f.write(f"{concept}|{','.join(str(d) for d in logs)}|{streak}\n")

def load_progress():
    try:
        with open("tracker.txt", "r") as f:
            for line in f:
                parts = line.strip().split("|")
                if len(parts) == 3:
                    concept = parts[0]
                    logs = [date.fromisoformat(d) for d in parts[1].split(",") if d]
                    streak = int(parts[2])
                    concepts.append(concept)
                    practice_logs[concept] = logs
                    streaks[concept] = streak
    except FileNotFoundError:
        pass  


def add_concept():
    global concepts
    concept = input("Enter concept name: ")
    concepts.append(concept)
    print(f" Concept '{concept}' added successfully!")


def log_practice():
    if len(concepts)==0:
        print("No concepts added yet. Please add one first.")
    else:
        print("Available concepts:", concepts)
        concept = input("Which concept did you practice? ")
        if concept in concepts:
            today = date.today()
            if concept not in practice_logs:
                practice_logs[concept] = []
                streaks[concept] = 0

            # Add today's log
            practice_logs[concept].append(today)

            # Check streak
            if len(practice_logs[concept]) > 1:
                last_date = practice_logs[concept][-2]
                if today == last_date + timedelta(days=1):
                    streaks[concept] += 1
                elif today == last_date:
                    pass  # same day, streak unchanged
                else:
                    streaks[concept] = 1
            else:
                streaks[concept] = 1

            print(f"Logged practice for '{concept}' today!")
        else:
            print(" Concept not found. Please add it first.")
        print(f"Current streak for {concept}: {streaks[concept]} days")
    save_progress() 


def view_stats():
    print("Concepts tracked so far:", concepts)
    for concept in concepts:
        streak = streaks.get(concept, 0)
        total_logs = len(practice_logs.get(concept, []))
        print(f" {concept}: Streak = {streak} days, Total logs = {total_logs}")

load_progress()
```

```python
import random

def number_guessing():
    number = random.randint(1,100)
    guess = None
    while guess != number:
        user_input = input("Enter your guess (1-100) or 'q' to quit: ")
        if user_input.lower() == "q":
            print(" Exiting Number Guessing game...")
            return
        try:
            guess = int(user_input)
        except ValueError:
            print(" Invalid input! Please enter a number or 'q'.")
            continue

        if guess < number:
            print("Too low!")
        elif guess > number:
            print("Too high!")
    print("Correct! You guessed it!")
    save_progress()
```

```python
def tic_tac_toe():
    board = [" "]*9

    def print_board():
        print(board[0]+"|"+board[1]+"|"+board[2])
        print("-+-+-")
        print(board[3]+"|"+board[4]+"|"+board[5])
        print("-+-+-")
        print(board[6]+"|"+board[7]+"|"+board[8])

    def check_winner(player):
        wins = [(0,1,2),(3,4,5),(6,7,8),
                (0,3,6),(1,4,7),(2,5,8),
                (0,4,8),(2,4,6)]
        return any(board[a]==board[b]==board[c]==player for a,b,c in wins)

    player = "X"
    for turn in range(9):
        print_board()
        move = input(f"Player {player}, enter position (1-9) or 'q' to quit: ")
        if move.lower() == "q":
            print(" Exiting Tic-Tac-Toe game...")
            return
        try:
            move = int(move) - 1
        except ValueError:
            print(" Invalid input! Please enter a number or 'q'.")
            continue

        if 0 <= move < 9 and board[move] == " ":
            board[move] = player
            if check_winner(player):
                print_board()
                print(f" Player {player} wins!")
                return
            player = "O" if player == "X" else "X"
        else:
            print("Invalid move! Try again.")
    print_board()
    print("It's a draw!")
    save_progress()

```

```python

def print_board(board):
    for i in range(9):
        if i % 3 == 0 and i != 0:
            print("------+-------+------")
        row = ""
        for j in range(9):
            if j % 3 == 0 and j != 0:
                row += "| "
            row += str(board[i][j]) if board[i][j] != 0 else "."
            row += " "
        print(row)

def is_valid(board, row, col, num):
    if num in board[row]:
        return False
    for i in range(9):
        if board[i][col] == num:
            return False
    start_row, start_col = (row//3)*3, (col//3)*3
    for i in range(3):
        for j in range(3):
            if board[start_row+i][start_col+j] == num:
                return False
    return True


def sudoku():
    board = [
        [0,0,3,0,2,0,6,0,0],
        [9,0,0,3,0,5,0,0,1],
        [0,0,1,8,0,6,4,0,0],
        [0,0,8,1,0,2,9,0,0],
        [7,0,0,0,0,0,0,0,8],
        [0,0,6,7,0,8,2,0,0],
        [0,0,2,6,0,9,5,0,0],
        [8,0,0,2,0,3,0,0,9],
        [0,0,5,0,1,0,3,0,0]
    ]

    while True:
        print_board(board)
        row = input("Enter row (1-9) or 'q' to quit: ")
        if row.lower() == "q":
            print(" Exiting Sudoku game...")
            break
        col = input("Enter column (1-9) or 'q' to quit: ")
        if col.lower() == "q":
            print(" Exiting Sudoku game...")
            break
        num = input("Enter number (1-9) or 'q' to quit: ")
        if num.lower() == "q":
            print(" Exiting Sudoku game...")
            break

        try:
            row, col, num = int(row)-1, int(col)-1, int(num)
        except ValueError:
            print(" Invalid input! Please enter numbers only.")
            continue

        if board[row][col] != 0:
            print(" Cell already filled!")
            continue

        if is_valid(board, row, col, num):
            board[row][col] = num
            print(" Move accepted!\n")
        else:
            print(" Invalid move! Try again.\n")

        if all(all(cell != 0 for cell in row) for row in board):
            print_board(board)
            print(" Congratulations! You solved the Sudoku puzzle!")
            print("Logged: Algorithms practiced today")
            break
            save_progress()
```

```python
while True:
    print("\n--- Learning Tracker Menu ---")
    print("1. Add Concept")
    print("2. Log Practice")
    print("3. View Stats")
    print("4. Play Game")
    print("5. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        add_concept()
    elif choice == "2":
        log_practice()
    elif choice == "3":
        view_stats()
    elif choice == "4":
        print(" Choose a game:")
        print("a) Number Guessing")
        print("b) Tic-Tac-Toe")
        print("c) Sudoku")
        game_choice = input("Enter choice (a/b/c): ")
        if game_choice == "a":
            number_guessing()
        elif game_choice == "b":
            tic_tac_toe()
        elif game_choice == "c":
            sudoku()
        else:
            print(" Invalid game choice.")
    elif choice == "5":
        print(" Exiting... Progress saved.")
        break
    else:
        print(" Invalid choice. Please enter 1-5.")
```
